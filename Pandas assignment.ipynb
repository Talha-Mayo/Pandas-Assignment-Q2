```python
# reading all csv files by applying glob and loop

census_files = glob.glob('Pandas Assignment/states*.csv')

data_list = []
for _file in census_files:
    data = pd.read_csv(_file)
    data_list.append(data)


#census_files
```

```python
# Concatinate the lists in 'data_list' by applying 'pd.concat()' to convert in into a pandas dataframe

us_census = pd.concat(data_list, ignore_index=True)
```

```python
us_census.head()
```

```python
#Checking the info of data frame

us_census.info()
```

```python
# Drop the column named "Unnamed: 0" bcz it is not required in data analysis

us_census=us_census.drop('Unnamed: 0',axis=1)
us_census.head()
```

```python
# get the info [head()] about the "Income" column
us_census['Income'].head()
```

```python
# Remove the "$" symbol from the data in 'Income' column

us_census['Income'] = us_census['Income'].replace('[$]','',regex=True)
```

```python
# Convert Income into numeric value

us_census['Income'] = pd.to_numeric(us_census['Income'])
```

```python

us_census['Income'].head()
```

```python
#Checking data type of Income colums
us_census['Income'].dtype
```

```python
# Before spliting 'GenderPop' we have to see the values inside it.

us_census.GenderPop.head()
```

```python
# Spliting 'GenderPop' column into 'Men' and 'Women' columns

us_census['Men'] = us_census['GenderPop'].str.split('(_)', expand=True)[0]
us_census['Women'] = us_census['GenderPop'].str.split('(_)', expand=True)[2]
us_census.head()
```

```python
# To convert object type data into numeric in 'Men' column first we have to Remove the 'M' from values

us_census['Men'] = us_census['Men'].replace('M','',regex=True)
```

```python
#Converting Object of column 'Men' values to numeric values

us_census['Men'] = pd.to_numeric(us_census.Men)

us_census.Men.head()
```

```python
# To convert object type data into numeric in 'Women' column first we have to Remove the 'F' from values

us_census['Women'] = us_census['Women'].replace('F','',regex=True)
```

```python
#Converting Object of column 'Women' values to numeric values

us_census['Women'] = pd.to_numeric(us_census['Women'])

us_census['Women'].head()
```

```python
us_census.head()
```

```python
# Now plot (scatter plot) the women's income data 

plt.scatter(us_census.Women, us_census.Income)
plt.xlabel('Women')
plt.ylabel('Income')
plt.show()
```

```python
# To check Null Value Counts

us_census.Women.isnull().value_counts()
```

```python
# According to the condition given in assignment we have to fill null values in women column with (Total Population - Men's column)

us_census['Women'] = us_census['Women'].fillna(us_census['TotalPop'] - us_census['Men'])
```

```python
inventory['in_stock'] = inventory.apply(lambda row: True if row.quantity > 0 else False, axis=1)

inventory.head()
```

```python
inventory['total_value'] = inventory.apply(lambda row: row.price * row.quantity, axis=1)

inventory.head()
```

```python
combine_lambda = lambda row: '{} - {}'.format(row.product_type, row.product_description)

combine_lambda
```

```python
inventory['full_description'] = inventory.apply(combine_lambda, axis=1)
```
